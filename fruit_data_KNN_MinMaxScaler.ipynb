```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
fruits = pd.read_table('fruit_data_with_colors.txt')
```

```python
fruits.head()
```

```python
fruits
```

```python
##spliting the data in training and testing datasets

X = fruits[['height', 'width', 'mass', 'color_score']]

Y = fruits['fruit_label']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=0)
```

```python
X_train
```

```python
X_test
```

```python
## Using normalization feature with MinMax scaling

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
```

```python
## fitting the scaler feature on the training set

scaler.fit(X_train)
```

```python
## store the transformed one in the new varible

X_train_scaled = scaler.transform(X_train) 

X_test_scaled = scaler.transform(X_test)
```

```python
## using KNN on unscaled dataset
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train, Y_train)
```

```python
## Accuracy of K-NN classifier on unscaled training set
knn.score(X_train,Y_train)
```

```python
## Accuracy of K-NN classifier on unscaled testing set
knn.score(X_test,Y_test)
```

```python
## using KNN on scaled dataset
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train_scaled, Y_train)
```

```python
##  Accuracy of K-NN classifier on scaled training set

knn.score(X_train_scaled, Y_train)
```

```python
## Accuracy of K-NN classifier on scaled testing set

knn.score(X_test_scaled, Y_test)
```

```python
## Result with the help of MinMaxScaler we are able to increase the accuray from 53.33% 
```
