```python
def split_data():
    winedf = creating_df()
    
    X = winedf
    Y = winedf.pop('target')
    
    return X, Y
```

```python
from sklearn.model_selection import train_test_split

def train_test():
    X,Y = split_data()
    
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=0)
    
    return X_train, X_test, Y_train, Y_test

X_train, X_test, Y_train, Y_test = train_test()
```
