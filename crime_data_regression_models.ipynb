```python
## apply the KNN, Linear, Ridge and Lasso regresion models on the crime dataset
```

```python
from adspy_shared_utilities import load_crime_dataset
```

```python
(X_crime, Y_crime) = load_crime_dataset()
```

```python
X_crime
```

```python
Y_crime
```

```python
## splitting the dataset into training and testing datasets
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_crime, Y_crime, random_state = 0)
```

```python
## apply KNN Regression to train the model
from sklearn.neighbors import KNeighborsRegressor
```

```python
knnreg = KNeighborsRegressor(n_neighbors = 5)
```

```python
knnreg.fit(X_train, Y_train)
```

```python
## accuracy rate on training dataset
knnreg.score(X_train, Y_train)
```

```python
## accuracy rate on testing dataset
knnreg.score(X_test, Y_test)
```

```python
## apply Linear Regression Model 
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
```

```python
linreg.fit(X_train, Y_train)
```

```python
linreg.coef_
```

```python
linreg.intercept_
```

```python
## accuracy rate of linear_regression model on training dataset
linreg.score(X_train, Y_train)
```

```python
## accuracy rate of linear_regression model on testing dataset
linreg.score(X_test, Y_test)
```

```python
## apply the Ridge Regression Model
from sklearn.linear_model import Ridge
linridge = Ridge(alpha=20.0)
```

```python
linridge.fit(X_train, Y_train)
```

```python
linridge.coef_
```

```python
linridge.intercept_
```

```python
## accuracy score on training dataset
linridge.score(X_train, Y_train)
```

```python
linridge.score(X_test, Y_test)
```

```python
## scaled the feature using MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
linridge_scaler = Ridge(alpha = 20.0)
```

```python
linridge_scaler.fit(X_train_scaled, Y_train)
```

```python
linridge_scaler.coef_
```

```python
linridge_scaler.intercept_
```

```python
## accuracy on training dataset
linridge_scaler.score(X_train_scaled, Y_train)
```

```python
## accuracy score on the testing dataset
linridge_scaler.score(X_test_scaled, Y_test)
```

```python
## apply the Lasso Regression model
from sklearn.linear_model import Lasso
```

```python
linlasso_scaled = Lasso(alpha = 2.0, max_iter = 10000)
```

```python
linlasso_scaled.fit(X_train_scaled, Y_train)
```

```python
linlasso_scaled.coef_
```

```python
linlasso_scaled.intercept_
```

```python
## accuracy score of lasso model along with MinMax Scaler feature on training dataset
linlasso_scaled.score(X_train_scaled, Y_train)
```

```python
## accuracy score of lasso model along with MinMax Scaler feature on testing dataset
linlasso_scaled.score(X_test_scaled, Y_test)
```

```python
## Result
# Using KNN Regressor Model accuracy on :
### Training dataset = 51.3%
### Testing dataset = 33.4%

#Using Linear Regression model accuracy on :
### Training dataset = 67.3%
### Testing dataset = 49.55%

#Using Ridge Regression Model without scaling, accuracy on :
### Training dataset = 67.09%
### Testing dataset = 49.4%

## Using Regression model with MinMaxScaler feature, accuracy on :
### Training dataset = 61.4%
### Testing dataset = 59.8%

# Using Lasso Regression model with MinMaxScaler feature, accuracy on :
### Training dataset = 63%
### Testing dataset = 62.3%
```
